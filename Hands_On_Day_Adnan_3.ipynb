# logika if ... else ...

```python
lampu = "nyala"
if lampu == 'nyala':
  print("lampu nyala")
else:
  print("lampu mati")
```

# logika if ... elif ... else ...

```python
lampu = "hijau"
if lampu == 'nyala':
  print("lampu nyala")
elif lampu == 'hijau':
  print("lampu hijau")
else:
  print("lampu mati")
```
