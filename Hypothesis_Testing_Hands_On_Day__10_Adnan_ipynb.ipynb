```python
import pandas as pd #Python for Data Analysis
from scipy.stats import ttest_ind #SCIENTIFIC PYTHON(scipy)
from scipy.stats import chisquare #scientific python for testing using chi square
from scipy.stats import f_oneway #scientific python for testing using anova one way
```

```python
# Nilai ujian siswa
# H0: Rata-rata nilai kelas konvensional = kelas digital
# H1: Rata-rata nilai kelas digital > kelas konvensional
kelas_konvensional = [65, 70, 72, 68, 75, 74, 71, 69, 73, 70]
kelas_digital = [75, 78, 80, 77, 82, 79, 76, 81, 78, 80]
```

```python
t_stat, p_value = ttest_ind(
    kelas_digital,
    kelas_konvensional,
    # alternative='greater'
)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python
alpha = 0.05
```

```python
# Jika p-value < 0,05, maka dapat disimpulkan bahwa metode pembelajaran digital interaktif secara statistik meningkatkan rata-rata nilai siswa dibandingkan metode konvensional.
if p_value < alpha:
    print("Tolak H0: Metode digital meningkatkan nilai siswa")
else:
    print("Gagal menolak H0: Tidak ada bukti peningkatan nilai")
```

```python
data = {
    "Metode": ['A','A','A','B','B','B','C','C','C'],
    "Nilai": [78,80,85,70,73,75,90,92,91],

}

df = pd.DataFrame(data)
df
```

```python
df.groupby("Metode")["Nilai"].mean()
```

```python
nilai_a = df[df["Metode"]== "A"]["Nilai"]
nilai_b = df[df["Metode"]== "B"]["Nilai"]
nilai_c = df[df["Metode"]== "C"]["Nilai"]
```

```python
nilai_c
```

```python
f_stat, p_value = f_oneway(nilai_a,nilai_b,nilai_c)
```

```python
print("F-Statistic:", f_stat)
print("P-Value:", p_value)
```
