#Soal 1: Membuat Class dan Object
Buatlah class dan object dengan ketentuan sebagai berikut:
1. Sebuah class bernama Mahasiswa
2. Tambahkan tiga atribut di dalam class tersebut, yaitu: nama, nim, dan jurusan.
3. Buatlah sebuah method/function bernama info_mahasiswa yang berfungsi untuk menampilkan informasi mahasiswa (nama, nim, dan jurusan).


* Perhatikan bagaimana mendeklarasikan class dan mendefinisikan atribut serta metode.
* Pastikan atribut dapat diakses dalam class dengan benar.
* Gunakan metode untuk menampilkan informasi dari instance yang dibuat.

Goals: Memahami konsep dasar class dan object, serta bagaimana cara mengakses atribut dan metode dalam class.

```python
class Mahasiswa:
  def __init__(self, nama, nim, jurusan):
    self.nama = nama
    self.nim = nim
    self.jurusan = jurusan

  def info_mahasiswa(self):
    print("Nama Mahasiswa : " , self.nama)
    print("NIM Mahasiswa  : " , self.nim)
    print("Jurusan        : " , self.jurusan)
```

```python
m1 = Mahasiswa('Andi', 12345, 'Teknik Informatika')
```

```python
m1.info_mahasiswa()
```

# Soal 2: Mengimplementasikan Encapsulation
1. Buatlah sebuah class bernama Tabungan.
2. Tambahkan sebuah atribut saldo yang bersifat private di dalam class tersebut.
3. Buatlah method deposit yang berfungsi menambah saldo sesuai jumlah yang diberikan.
4. Buatlah method tarik untuk mengurangi saldo.
5. Berikan pengecekan di dalam metode tarik sehingga:
* Jika saldo tidak mencukupi, tampilkan pesan: "Saldo tidak mencukupi."
* Jika saldo mencukupi, proses pengurangan saldo dapat dilakukan.
6. Ujilah class yang kamu buat dengan membuat minimal satu object Tabungan, kemudian panggil metode deposit dan tarik untuk melihat apakah fungsinya berjalan sesuai ketentuan.


* Pahami konsep aksesibilitas atribut dalam OOP.
* Pastikan pengelolaan atribut private dapat dilakukan melalui metode yang sesuai.
* Terapkan validasi yang diperlukan untuk menjaga integritas data.


```python
class Tabungan:
  def __init__(self, saldo):
    # private
    self.__saldo = saldo

  def deposit(self,tambahsaldo):
    self.__saldo += tambahsaldo

  def tarik(self,pengurangansaldo):
    if(self.__saldo < pengurangansaldo):
      print("saldo tidak cukup")
    else:
      self.__saldo -= pengurangansaldo

  def getSaldo(self):
    return self.__saldo

  def cetak(self):
    print(self.getSaldo())
```

```python
T1 = Tabungan(5000)
```

```python
T1.deposit(10000)
```

```python
T1.cetak()
```

```python
T1.tarik(5000)
```

```python
T1.cetak()
```

```python
T1.tarik(20000)
```

# Soal 3: Mengimplementasikan Inheritance
1. Buatlah sebuah class bernama Pegawai.
2. Tambahkan dua atribut di dalam class Pegawai, yaitu: nama dan gaji.
3. Buatlah sebuah subclass bernama Manager yang mewarisi class Pegawai.
4. Tambahkan sebuah atribut baru pada class Manager, yaitu tunjangan.
5. Buatlah metode info_manager pada class Manager yang berfungsi menampilkan informasi nama, gaji, dan tunjangan dari instance Manager.
6. Pastikan subclass Manager dapat mengakses atribut yang berasal dari class induk (Pegawai).
7. Ujilah class Manager dengan membuat minimal satu instance Manager, kemudian panggil metode info_manager.



* Pahami bagaimana inheritance memungkinkan penggunaan kembali kode dari superclass.
* Pastikan subclass dapat memperluas fungsionalitas superclass dengan atribut/metode tambahan.
* Pastikan relasi antara superclass dan subclass terimplementasi dengan baik.


```python
class Pegawai:
  def __init__(self,nama,gaji):
    self.nama = nama
    self.gaji = gaji
```

```python
class Manager(Pegawai):
  def __init__(self,nama,gaji,tunjangan):
    # kunci pewarisan di super
    super().__init__(nama,gaji)
    self.tunjangan = tunjangan

  def info_manager(self):
    print("Nama      : " , self.nama)
    print("Gaji      : " , self.gaji)
    print("tunjangan : " , self.tunjangan)
```

```python
M1 = Manager("jono",30000,5000)
```

```python
M1.info_manager()
```

# Soal 4: Polymorphism dan Method Overriding
1. Buatlah sebuah class bernama Kendaraan.
2. Tambahkan sebuah method bernama bahan_bakar di dalam class Kendaraan yang hanya menampilkan pesan:
 "Jenis bahan bakar tidak diketahui."
3. Buatlah subclass bernama Mobil yang mewarisi class Kendaraan.
4. Buatlah subclass bernama Motor yang juga mewarisi class Kendaraan.
5. Override method bahan_bakar pada class Mobil sehingga menampilkan jenis bahan bakar yang digunakan (misalnya "Bensin").
6. Override method bahan_bakar pada class Motor untuk menampilkan jenis bahan bakarnya (misalnya "Solar", atau sesuai arahan soal).
7. Ujilah ketiga class tersebut dengan membuat instance dari Kendaraan, Mobil, dan Motor, kemudian panggil method bahan_bakar pada masing-masing object untuk melihat perbedaannya.


* Pahami konsep method overriding dan bagaimana menerapkannya dalam subclass.
* Pastikan method yang di-override memberikan hasil yang sesuai dengan subclass terkait.
* Gunakan metode yang sama tetapi dengan implementasi berbeda untuk setiap subclass.


```python
class Kendaraan:
  def bahan_bakar(self):
    print("jenis bahan bakar tidak diketahui")
```

```python
class Mobil(Kendaraan):
  def bahan_bakar(self):
    print("Bensin")
```

```python
class Motor(Kendaraan):
  def bahan_bakar(self):
    print("Solar")
```

```python
k1 = Kendaraan()
mobil1 = Mobil()
motor1 = Motor()
```

```python
k1.bahan_bakar()
mobil1.bahan_bakar()
motor1.bahan_bakar()
```

# Soal 5: Abstraction dan Interfaces
1. Buatlah sebuah class abstrak bernama Hewan.
2. Tambahkan sebuah method abstrak bernama suara di dalam class Hewan.
Notes: Method ini harus dideklarasikan sebagai abstrak, tanpa implementasi.
3. Buatlah dua buah subclass bernama Kucing dan Anjing yang mewarisi class Hewan.
4. Implementasikan method suara pada class Kucing dan Anjing sehingga menghasilkan suara khas hewan tersebut (contoh: "Meow" untuk kucing, “Bark” untuk anjing).
5. Ujilah kedua subclass dengan membuat instance dari Kucing dan Anjing, lalu panggil method suara untuk memastikan implementasi berjalan sesuai ketentuan.
* Pahami konsep abstraction dalam OOP dan bagaimana class abstrak digunakan.
* Implementasikan metode abstrak yang diwajibkan pada subclass.
* Pastikan kode tetap terstruktur dan mengikuti prinsip OOP


```python
from abc import ABC,abstractmethod
class Hewan(ABC):
  def suara(self):
    pass
```

```python
class Kucing(Hewan):
  def suara(self):
    print("Meow")
```

```python
class Anjing(Hewan):
  def suara(self):
    print("Bark")
```

```python
kucing_1 = Kucing()
anjing_1 = Anjing()
```

```python
kucing_1.suara()
anjing_1.suara()
```
