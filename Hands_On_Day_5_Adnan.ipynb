# Loop DataFrame

## Iterrows

```python
# Untuk simulasi, kita ambil 5 baris saja secara acak.
df_sampel = df.sample(5)

for index, row in df_sampel.iterrows():
  # Mengeluarkan nilai dari kolom "Title" dan kolom "location".
  print(index, "| ", row["Title"], " | ", row["location"])
```

```python
# Kita coba memodifikasi data sampel tsb, mengganti "location" agar menjadi
# huruf besar semua.
for index, row in df_sampel.iterrows():
  # Menggunakan method .upper() untuk kapitalisasi "location".
  print(row["location"].upper())
```
