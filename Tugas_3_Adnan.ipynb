# Soal 2: Membuat input
Buatlah dua kode agar pengguna bisa melakukan input untuk data berikut:
* Nama produk
* Jumlah produk terjual
* Harga produk


```python
#input nama produk
nama_produk = input("Masukkan Nama Produk : ")

#input jumlah produk terjual
jml_produk = int(input("Jumlah Produk Terjual : "))

#input harga produk
harga_produk = int(input("Harga Produk : "))
```

# Soal 3: Logika If/Else untuk Rekomendasi Produk
## Lakukan perhitungan total revenue dari penjualan barang dengan rumus:
**total revenue = jumlah_produk_terjual  x harga_produk.**
## Gunakan conditional statement untuk melihat hasil akhir penilaian:
* Jika total revenue kurang dari atau sama dengan 2.000.000, maka kategori penjualan “Rendah”.
* Jika total revenue lebih dari 2.000.000 dan total revenue kurang dari atau sama dengan  8.000.000, maka kategori penjualan “Sedang”.
* Jika total revenue lebih dari 8.000.000, maka kategori penjualan “Tinggi”.


```python
total_revenue = jml_produk * harga_produk

#jika total revenue kurang dari 2.000.000
if total_revenue <= 2000000:
  kategori = "kategori penjualan rendah"

#jika total revenue lebih dari 2.000.000 dan kurang dari sama dengan 8.000.000
elif total_revenue > 2000000 and total_revenue <= 8000000:
  kategori = "kategori penjualan sedang"

#jika total revenue lebih dari 8.000.000
else:
  kategori = "kategori penjualan tinggi"
```

# Soal 4: Output Akhir Program
## Tampilkan rekomendasi hasil analisis penjualan yang meliputi:
* Nama Produk
* Jumlah Terjual
* Harga Produk
* Total Revenue
* Kategori Penjualan (Rendah, sedang, tinggi)


```python
#menampilkan rekomendasi hasil analisis penjualan
print("nama produk   : " , nama_produk)
print("jumlah produk : " , jml_produk)
print("harga produk  : " , harga_produk)
print("total revenue : " , total_revenue)
print("rekomendasi produk : " , kategori)
```
