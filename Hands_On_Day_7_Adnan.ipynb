```python
# Menggunakan template dari LeetCode
class Solution:
    def scoreOfString(self, s: str) -> int:
        jumlah = 0

        # len(s) adalah panjang kata
        # range(panjang kata) menghasilkan indeks 0, 1, 2, ..., panjang kata - 1
        for i in range(len(s)-1):
            pairs = s[i:i+2] # misalnya, he

            char_pertama = pairs[0] # ngambil h-nya aja
            char_kedua = pairs[1]   # ngambil e-nya aja

            ascii_pertama = ord(char_pertama) # ubah h menjadi 104
            ascii_kedua = ord(char_kedua)     # ubah e menjadi 101

            selisih = abs(ascii_pertama - ascii_kedua)
            jumlah += selisih

        return jumlah
```
