Membuat property menjadi private dengan __.


```python
class Kucing():
 def __init__(self, nama):
  self.nama_publik = nama
  self.__nama_privat = nama ## private
k1 = Kucing("Luna")
k1.nama_publik = "Luna029"
k1.__nama_privat = "lunasdf" # error
print("Nama publik:", k1.nama_publik)
print("Nama privat:", k1.__nama_privat) # error
```

```python
# kode ini pemisalan untuk mengecek apakah nama itu huruf semua dan tidak asdf
# hanya gambaran
class Kucing():
 def __init__(self, nama):
  self.nama_publik = nama
  self.__nama_privat = nama
 def get_nama_privat(self):
  return self.__nama_privat
 def set_nama_privat(self, nama):
  if nama huruf semua and tidak asdf:
    self.__nama_privat = nama
  else:
    print("Tidak boleh ganti nama.")
k1 = Kucing("Luna")
k1.set_nama_privat("Luna029")
k1.set_nama_privat("Artemis")
print("Nama privat:", k1.get_nama_privat())
```

# Fungsi range() vs enumerate()
Mengiterasi indeks dan elemen dari suatu list.


```python
# Biasa
elemens = ["kereta api", "kapal", "kuda"]
# range menerima input angka integer,
# yakni banyaknya elemen di list elemens.
for i in range(len(elemens)):
 # mengambil elemen ke-i dengan sintaksis
 # elemens[i]
 print(f"Indeks {i} elemen {elemens[i]}")
```

# Inheritance

```python
class Character:

  def __init__(self, damage, health, bleed):
    self.damage = damage

    # Membuat property health menjadi private
    # agar tidak seenaknya diubah-ubah.
    self.__health = health

    self.__bleed = bleed

  def get_health(self):
    return self.__health

  def attack(self):
    return self.damage

  def attacked(self):
    self.__health -= self.__bleed
```

```python
class Hero(Character):

  def __init__(self, damage, health, bleed):
    super().__init__(damage, health, bleed)
```

```python
class Enemy(Character):

  def __init__(self, damage, health, bleed):
    super().__init__(damage, health, bleed)
```

# Polymorphism

```python
class Character:

  def __init__(self, damage, health, bleed):
    self.damage = damage

    # Membuat property health menjadi private
    # agar tidak seenaknya diubah-ubah.
    self.__health = health

    self.__bleed = bleed

  def get_health(self):
    return self.__health

  def attack(self):
    return self.damage

  def powerup(self):
    # belum tahu berapa besar damage dari hasil powerup
    # karena bergantung apakah dia hero atau enemy
    return 0

  def attacked(self):
    self.__health -= self.__bleed
```

```python
class Hero(Character):

  def __init__(self, damage, health, bleed):
    super().__init__(damage, health, bleed)

  def powerup(self):
    return self.damage + 1
```

```python
class Enemy(Character):

  def __init__(self, damage, health, bleed):
    super().__init__(damage, health, bleed)

  def powerup(self):
    return self.damage
```

```python
h = Hero(2, 20, 2)
h.powerup()
```

```python
e = Enemy(1, 10, 3)
e.powerup()
```

# Abstraction

```python
from abc import ABC, abstractmethod
```

```python
class Character(ABC):

  def __init__(self, damage, health, bleed):
    self.damage = damage

    # Membuat property health menjadi private
    # agar tidak seenaknya diubah-ubah.
    self.__health = health

    self.__bleed = bleed

  def get_health(self):
    return self.__health

  def attack(self):
    return self.damage

  @abstractmethod
  def powerup(self):
    # belum tahu berapa besar damage dari hasil powerup
    # karena bergantung apakah dia hero atau enemy
    pass

  def attacked(self):
    self.__health -= self.__bleed
```

```python
class Hero(Character):

  def __init__(self, damage, health, bleed):
    super().__init__(damage, health, bleed)

  def powerup(self):
    return self.damage + 1
```

```python
class Enemy(Character):

  def __init__(self, damage, health, bleed):
    super().__init__(damage, health, bleed)

  def powerup(self):
    return self.damage
```

```python
c = Character(2, 20, 2)
```

```python
h = Hero(2, 20, 2)
h.powerup()
```

```python
e = Enemy(1, 10, 3)
e.powerup()
```
